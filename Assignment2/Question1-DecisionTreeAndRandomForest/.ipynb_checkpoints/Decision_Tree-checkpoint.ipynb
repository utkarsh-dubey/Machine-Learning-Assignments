```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset.csv')
```

```python
x = df.dropna()
```

```python
x = pd.get_dummies(x)
x.head()
```

```python
def train_test_split(dataset:pd.DataFrame, split=0.7, seed=0):
    columns = dataset.columns
    dataset = dataset.to_numpy()
    np.random.seed(seed)
    train = []
    train_size = split * len(dataset)
    dataset_copy = list(dataset)
    while len(train) < train_size:
        index = np.random.randint(len(dataset_copy))
        train.append(dataset_copy.pop(index))
    train = pd.DataFrame(train, columns=columns)
    dataset_copy = pd.DataFrame(dataset_copy, columns=columns)
    return train, dataset_copy

def precision(label, confusion_matrix):
    # precision = TP/(TP+FP)
    col = confusion_matrix[:, label]
    return confusion_matrix[label, label] / col.sum()
    
def recall(label, confusion_matrix):
    # recall = TP/(TP+FN)
    row = confusion_matrix[label, :]
    return confusion_matrix[label, label] / row.sum()

def accuracy(confusion_matrix):
    # accuracy = TP + TN
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements 

def precision_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_precisions = 0
    for label in range(rows):
        sum_of_precisions += precision(label, confusion_matrix)
    return sum_of_precisions / rows

def recall_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_recalls = 0
    for label in range(columns):
        sum_of_recalls += recall(label, confusion_matrix)
    return sum_of_recalls / columns

def accuracy(confusion_matrix):
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements 

def f1Score(confusion_matrix):
    pre = precision_macro_average(confusion_matrix)
    recal = recall_macro_average(confusion_matrix)
    return 2*(pre*recal)/(pre+recal)

def printStat(cm):
    print("precision", precision_macro_average(cm))
    print("recall", recall_macro_average(cm))
    print("accuracy", accuracy(cm))
    print("f1 Score", f1Score(cm))
  

def plotAccuracy(model, trainX, trainY, testX, testY):
    plt.figure(figsize=(20, 6))
    plt.subplot(1, 2, 1)
    y_pred = model.predict(testX)
    y_train_pred = model.predict(trainX)
    plt.scatter(testY, y_pred, c='blue')
    p1 = max(max(y_pred), max(testY))
    p2 = min(min(y_pred), min(testY))
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.xlabel('True Values (Test Data)', fontsize=15)
    plt.ylabel('Predictions (Test Data)', fontsize=15)
    plt.axis('equal')
    plt.title('Decision Tree Regression (Test Data)')

    plt.subplot(1, 2, 2)
    plt.scatter(trainY, y_train_pred, c='red')
    p1 = max(max(y_train_pred), max(trainY))
    p2 = min(min(y_train_pred), min(trainY))
    plt.plot([p1, p2], [p1, p2], 'b-')
    plt.xlabel('True Values (Training Data)', fontsize=15)
    plt.ylabel('Predictions (Training Data)', fontsize=15)
    plt.axis('equal')
    plt.title('Decision Tree Regression (Training Data)')

    plt.show()

def plotConfusionMatixAndMetrix(predY, testY):
    M = confusion_matrix(testY, predY)
    plt.figure(figsize = (15,10))
    sns.heatmap(M/np.sum(M), fmt='.2%', annot=True)
    printStat(M)
    return M
```

```python
train, val_test = train_test_split(x)
validation, test = train_test_split(val_test, split=0.5)
trainX, trainY = train.drop('month', axis=1), train['month']
validationX, validationY = validation.drop('month', axis=1), validation['month']
testX, testY = test.drop('month', axis=1), test['month']
# mean = trainX.mean(axis=0)
# std = trainX.std(axis=0)
# trainX = (trainX-mean)/std
# testX = (testX-mean)/std
```

# Forest

```python
class Forest:
    def __init__(self, forest_size=100, **kwargs):
        self.forest = []
        self.forest_size = forest_size
        for tree in range(self.forest_size):
            self.forest.append(DecisionTreeClassifier(**kwargs))
          
    def fit(self, train, output_column):
        for tree in range(self.forest_size):
            data = train.sample(frac=0.50, random_state=tree)
            self.forest[tree].fit(data.drop(output_column, axis=1), data[output_column])
        return self
          
    def predict(self, testX):
        prediction = [{} for row in range(len(testX))]
        for tree in self.forest:
            predY = tree.predict(testX)
            for index, predict in enumerate(predY):
                prediction[index][predict] = prediction[index].get(predict, 0) + 1
        return np.array([max(x, key=x.get) for x in prediction])
    
    def score(self, X, Y):
        return (self.predict(X)==Y).mean()
```

```python
forest = Forest(forest_size = 100, criterion='entropy', random_state=0, max_depth=3)
forest.fit(train, 'month')
forest.score(testX, testY)
```

```python
predY = forest.predict(testX)
```

```python
cm = confusion_matrix(testY, predY)
print("Accuracy: ", accuracy(cm))
plt.figure(figsize = (15,10))
sns.heatmap(cm/np.sum(cm), fmt='.2%', annot=True)
```

```python
v_forest_size_score = []
tr_forest_size_score = []
te_forest_size_score = []

for size in range(10, 101, 10):
    f = Forest(forest_size = size, criterion='entropy', random_state=0, max_depth=15).fit(train, 'month')
    v_forest_size_score.append(f.score(validationX, validationY))
    tr_forest_size_score.append(f.score(trainX, trainY))
    te_forest_size_score.append(f.score(testX, testY))
```

```python
plt.plot(range(10, 101, 10), v_forest_size_score, '-o', label='validation')
plt.plot(range(10, 101, 10), tr_forest_size_score, '-o', label='training')
plt.plot(range(10, 101, 10), te_forest_size_score, '-o', label='testing')
plt.legend()
plt.grid()
```

```python
f_forest = Forest(forest_size = 80, criterion='entropy', random_state=0, max_depth=15).fit(train, 'month')
```

```python
plotConfusionMatixAndMetrix(f_forest.predict(testX), testY)
```

# AdaBoost

```python
estimators = [4, 8, 10, 15, 20]
test_score = []
train_score = []
for estimates in estimators:
    clf = AdaBoostClassifier(n_estimators=estimates, random_state=0, base_estimator=DecisionTreeClassifier(random_state=0, criterion="entropy", max_depth=15))
    clf.fit(trainX, trainY)
    test_score.append(clf.score(testX, testY))
    train_score.append(clf.score(trainX, trainY))
```

```python
plt.plot(estimators, test_score, '-o', label='testing')
plt.plot(estimators, train_score, '-o', label='training')
plt.legend()
plt.grid()
```
